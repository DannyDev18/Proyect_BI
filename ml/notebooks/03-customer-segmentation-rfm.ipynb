### 1. Datos Sintéticos / Dummy para Simulación RFM
En producción, esta información vendrá de una consulta agrupando facturas desde la EDW.

```python
# Generando datos RFM representativos de 100 clientes ficticios
np.random.seed(42)
data_rfm = pd.DataFrame({
    'recency': np.random.randint(1, 365, size=100),       # Días transcurridos desde última compra
    'frequency': np.random.randint(1, 50, size=100),     # Cantidad total de transacciones
    'monetary': np.random.exponential(scale=500, size=100) # Gasto monetario total
})
data_rfm.index = [f'Cliente_{i}' for i in range(100)]
data_rfm.head()
```

### 2. Entrenamiento de K-Means

```python
# Entrenamos el pipeline con 3 clústeres
artifacts = train_rfm_segmentation(data_rfm, n_clusters=3)
model = artifacts['model']
scaler = artifacts['scaler']

# Asignar etiquetas a la data original
data_rfm['cluster'] = model.labels_
data_rfm.groupby('cluster').mean()
```

### 3. Visualización de Perfiles de Clientes

```python
sns.pairplot(data_rfm, hue='cluster', palette='Set2')
plt.suptitle('Segmentación K-Means RFM', y=1.02)
plt.show()
```
