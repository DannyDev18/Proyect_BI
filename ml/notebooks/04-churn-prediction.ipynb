```python
import pandas as pd
import numpy as np
from src.training.train_churn_prediction import train_churn_model, evaluate_churn_classifier
from sklearn.model_selection import train_test_split
```

### 1. Datos Dummy de Comportamiento de Clientes

```python
np.random.seed(42)
n_samples = 200

# Generando variables típicas de comportamiento
X = pd.DataFrame({
    'inactivity_days': np.random.randint(1, 180, n_samples),
    'average_ticket': np.random.uniform(10, 300, n_samples),
    'total_orders': np.random.randint(1, 40, n_samples),
    'discount_ratio': np.random.uniform(0.0, 0.4, n_samples)
})

# Definimos si abandonó basado en una regla probabilística heurística
churn_prob = 1 / (1 + np.exp(-(0.05 * X['inactivity_days'] - 0.01 * X['total_orders'] - 2)))
y = (churn_prob > np.random.uniform(0.0, 1.0, n_samples)).astype(int)

print(f"Tasa de Abandono (Churn Rate) en la muestra simulada: {y.mean() * 100}%")
```

### 2. Entrenamiento y Validación

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42, stratify=y)

model = train_churn_model(X_train, y_train, hyperparameter_search=True)

# Predecir resultados
y_pred = model.predict(X_test)
y_proba = model.predict_proba(X_test)

# Evaluar métricas de rendimiento
evaluate_churn_classifier(y_test, y_pred, y_proba)
```
